## 1 Средний бал и список отстающих учеников

```python
Иванов О. 4
Петров И. 3
Дмитриев Н. 2
Смирнова О. 4
Керченских В. 5
Котов Д. 2
Бирюкова Н. 1
Данилов П. 3
Аранских В. 5
Лемонов Ю. 2
Олегова К. 4
```

```python
# построчное заполнение словаря парами "фамилия-оценка"
studs = {}
start = '1'
while start == '1':
    studs.update({input('Фамилия: '):input('Оценка')})
    start = input('Продолжить? да-1, нет-0 ')
print(studs)
```

```python
def avg_mark(studs):
    summ = 0
    for mark in studs.values():
        summ += int(mark)
    avg_mark = summ / len(studs)
    print(f'Средний балл - {avg_mark}')
    print('Список отстающих учеников:')
    for last_name, mark in studs.items():
        if int(mark) < avg_mark:
            print(last_name)
            
print(avg_mark(studs))
```
